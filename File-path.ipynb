```python
import zipfile
import os

# Nama file zip
zip_file_path = "Indonesian License Plate Recognition Dataset.zip"

# Folder tujuan ekstraksi
extract_to = "Indonesian_LPR_Dataset"

# Cek apakah folder sudah ada dan tidak kosong
if os.path.exists(extract_to) and os.listdir(extract_to):
    print(f"Folder '{extract_to}' sudah berisi file. Ekstraksi dilewati.")
else:
    # Jika folder tidak ada atau kosong, lakukan ekstraksi
    if not os.path.exists(extract_to):
        os.makedirs(extract_to)

    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)
    print("Ekstraksi selesai ke folder:", extract_to)
```

```python
import os
import shutil
from pathlib import Path

# Path dasar
base_dir = Path("Indonesian_LPR_Dataset")

# Path folder sumber
splits = ["train", "test", "val"]

# Path folder tujuan
merged_dir = base_dir / "merged"

# Hapus folder merged jika sudah ada
if merged_dir.exists():
    shutil.rmtree(merged_dir)

# Buat ulang folder merged
for split in splits:
    (merged_dir / split).mkdir(parents=True, exist_ok=True)

# Proses copy file .jpg dan .txt dari masing-masing split
for split in splits:
    images_src = base_dir / "images" / split
    labels_src = base_dir / "labels" / split
    dst = merged_dir / split

    # Copy semua file .jpg
    for file in images_src.glob("*.jpg"):
        shutil.copy(file, dst / file.name)

    # Copy semua file .txt
    for file in labels_src.glob("*.txt"):
        shutil.copy(file, dst / file.name)

print("Penggabungan selesai.")
```

```python
# Tampilkan isi folder test setelah proses copy
print("\nIsi folder merged/test:")
test_path = merged_dir / "test"
for file in sorted(test_path.iterdir()):
    print("-", file.name)
```
